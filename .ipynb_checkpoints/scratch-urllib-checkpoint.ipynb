```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width: 100% !important; }</style>"))

import urllib.request
import json
#import numpy as np
from datetime import datetime,tzinfo,timezone,timedelta
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
%matplotlib inline
```

```python
def avgHourlyPrice(times, prices):
    nPrices = len(prices)
    rvTimes = times[nPrices-1::-1]
    rvPrices = prices[nPrices-1::-1]
    idx = 0
    tm = rvTimes[0]
    avgPrices = []
    hrEnds = []
    nextTm = tm+3600;

    while idx < (nPrices-1):
        tot = 0
        cnt = 0
        while tm<nextTm:
            tot += rvPrices[idx]
            cnt += 1
            idx += 1
            if idx == nPrices: break
            tm = rvTimes[idx]

        avgPrices.append(tot/cnt)
        hrEnds.append(nextTm)
        nextTm += 3600

    return avgPrices, hrEnds
```

```python
def utc_2_cst_str(millis):
    utcOffset = 0 # i think fromtimestamp assumes UTC and converts to local, so no offset?
    dttm = datetime.fromtimestamp(millis)+timedelta(hours=utcOffset)
    return '{}/{} {}:{:02}'.format(dttm.month,dttm.day, dttm.hour, dttm.minute)

startOffsetHrs = 36
endOffsetHrs = -8 #only returns historical numbers with most recent about 10minutes old...

feedType = '5minutefeed'
nowUtc = datetime.utcnow()
tmNow = nowUtc-timedelta(hours=endOffsetHrs, minutes=nowUtc.minute)
tmStrt = nowUtc-timedelta(hours=startOffsetHrs, minutes=(nowUtc.minute-60))

dateStart = '{}{:02}{:02}{:02}{:02}'.format(tmStrt.year,tmStrt.month,tmStrt.day,tmStrt.hour,tmStrt.minute)
dateEnd = '{}{:02}{:02}{:02}{:02}'.format(tmNow.year,tmNow.month,tmNow.day,tmNow.hour,tmNow.minute)


pricingUrl='https://hourlypricing.comed.com/api'
rUrl = '{0}?type={1}&format=json&datestart={2}&dateend={3}'.format(pricingUrl,feedType,dateStart,dateEnd)
# rUrl = '{0}?type={1}'.format(pricingUrl,feedType)
req=urllib.request.urlopen(rUrl).read()

if True: #figure out how to detect failure for urllib response
    rJson=json.loads(req)

    times = [int(elem['millisUTC'])//1000 for elem in rJson]
    prices = [float(elem['price']) for elem in rJson]
    
    avgPrices, endHrs = avgHourlyPrice(times, prices)
    pltHrs = [v-1800 for v in endHrs]

    fig = plt.figure(figsize=(28,12))
    ax = fig.add_subplot(111)
    
    xMinorSplit = AutoMinorLocator(3)
    yMinorSplit = AutoMinorLocator(4)
    ax.xaxis.set_minor_locator(xMinorSplit)
    ax.yaxis.set_minor_locator(yMinorSplit)
    
    #ax.minorticks_on()
    ax.grid(True, which='major', color='black')
    ax.grid(True, which='minor', color='gray', alpha=0.5)

    ax.plot(times, prices, color='blue', linestyle='-')
    ax.plot(pltHrs, avgPrices, color='red', linestyle='-')
    
    roundStep = 10800 #1hr
    tmStep = 10800 #3hrs
    tmStart = (times[-1]//roundStep)*roundStep
    tmEnd = ((times[0]+roundStep//2)//roundStep)*roundStep
    tmTicks = range(tmStart,tmEnd+1,tmStep)
    ax.set_xticks(tmTicks)
    ax.set_xticklabels([utc_2_cst_str(t) for t in tmTicks])
    
    ax.set_ylabel('/kWh',fontsize=20)

    plt.show()
else:
    print('failed')
```
